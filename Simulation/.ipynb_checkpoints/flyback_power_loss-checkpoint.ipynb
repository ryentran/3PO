```python
import numpy as np

import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
V_in = 12
V_out = 15

N_prim = 8
N_sec1 = 9
N_sec2 = 3

R_prim = 15e-3
R_sec1 = 22e-3
R_sec2 = 45e-3

R_ds = 4e-3

V_f = 0.31

R_L = 7.5

D = np.linspace(0, 0.99, 10000)
```

```python
R_out =  (((N_prim/N_sec1)**2)*(D**2/(1-D)**2)*(R_prim + R_ds) - R_sec1 - (V_f/V_out)*R_L)*D + R_sec1 + (V_f/V_out)*R_L
```

```python
R_out = R_out*1e3 # Converter to mOhms
```

```python
V_out_ideal = V_in*(D/(1-D))*(N_sec1/N_prim)*(R_L/(R_out/1e3+R_L))
```

```python
fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.02)
fig.add_trace(go.Scatter(x=D, y=R_out), row=1, col=1)
fig.add_trace(go.Scatter(x=D, y=V_out_ideal), row=2, col=1)

fig.update_xaxes(title_text='Duty Cycle', row=2, col=1)
fig.update_yaxes(title_text='R_out (mOhms)', row=1, col=1)
fig.update_yaxes(title_text='V_out (volts)', row=2, col=1)

fig.update_layout(
    height=1000,
    title='Duty Cycle vs Output Resistance',
    showlegend=False
)
```
